```python
import pandas as pd
```

```python
file_path = r"C:\Users\FREDERICK.S\Documents\ByteDataScience\Generated_Dataset.csv"
df = pd.read_csv(file_path)
```

```python
df.head()
```

```python
missing_values = df.isnull().sum()
```

```python
df_cleaned = df.drop_duplicates()
```

```python
df_cleaned.loc['Purchase_Date'] = pd.to_datetime(df_cleaned['Purchase_Date'], errors='coerce')
```

```python
df_cleaned = df_cleaned.dropna()
```

```python
df_cleaned['Quantity'] = df_cleaned['Quantity'].astype(int)
df_cleaned['Price_per_unit'] = df_cleaned['Price_per_unit'].astype(float)
```

```python
missing_values_after = df_cleaned.isnull().sum()
```

```python
missing_values, missing_values_after
```

```python
df_cleaned['Total_Revenue'] = df_cleaned['Quantity'] * df_cleaned['Price_per_unit']
total_revenue = df_cleaned['Total_Revenue'].sum
```

```python
top_5_products = df_cleaned['Product'].value_counts().head(5)
```

```python
unique_customers = df_cleaned['Customer_ID'].nunique()
```

```python
total_revenue, top_5_products, unique_customers
```

```python
import matplotlib.pyplot as plt
```

```python
plt.figure(figsize=(8, 5))
top_5_products.plot(kind='bar', color='gold', edgecolor='black')

plt.xlabel("Product")
plt.ylabel("Number of Purchases")
plt.title("Top 5 Selling Products")
plt.xticks(rotation=45)
plt.grid(axis='y', linestyle='--', alpha=0.7)

plt.show()
```
